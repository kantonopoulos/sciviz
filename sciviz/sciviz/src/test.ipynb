```python
import matplotlib.pyplot as plt
import seaborn as sns
from point import point
from histogram import histogram
from boxplot import boxplot
from bar import bar
from jitter import jitter
from sciviz import sciviz
from theme import theme
from violin import violin
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')

# Create the visualization with faceting by species
ax = sciviz(iris, height=5, aspect=1)
ax = violin(ax=ax, data=iris, y='species', x='sepal_width', color='species', split=False, orient='h', alpha=1, show_legend=True, box=True)

user_labels = {'Species': ['Setosa', 'Versicolor', 'Virginica']}
ax = theme(ax, 
           title="Iris Dataset", 
           subtitle="Sepal Length by Species", 
           title_size=14,
           yticks=[0, 1, 2],
           yticks_angle=90,
           yticks_labels=['setosa', 'Versicolor', 'Virginica'],
           theme="classic", 
           legend=True,
           legend_labels=user_labels, 
           legendtitles_size=11, 
           legendtitles_weight="bold", 
           legendlabels_size=10, 
           legendlabels_weight="normal",
           legend_pos="side")

plt.show()
```
