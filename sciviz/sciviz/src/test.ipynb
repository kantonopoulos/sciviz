```python
import matplotlib.pyplot as plt
import seaborn as sns
from point import point
from sciviz_main import sciviz
```

```python
iris = sns.load_dataset("iris")

# Create the visualization with faceting by species
ax = sciviz(iris, height=4, aspect=1)
point(iris, x="sepal_length", y="sepal_width", color="species", palette="jama", ax=ax, alpha = 0.8)
plt.show()
```

```python
iris = sns.load_dataset("iris")

# Create the visualization with faceting by species
ax = sciviz(iris, cols="species", height=4, aspect=1, col_wrap=3)
ax = point(iris, x="sepal_length", y="sepal_width", color="species", palette="deep", ax=ax)
```
