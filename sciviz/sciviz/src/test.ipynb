```python
iris = sns.load_dataset("iris")

# Create the visualization with faceting by species
ax = sciviz(iris, cols="species", height=4, aspect=1, col_wrap=3)
ax = point(ax=ax, data=iris, x="sepal_length", y="sepal_width", color="species", palette="jama")
ax = theme(ax, 
           title="Sepal Length vs Sepal Width",
           theme="classic", 
           xlab="Sepal Length 2", 
           ylab="Sepal Width 2",
           xlim=[3.5, 8.5],
           ylim=[1.5, 5],
           xticks=[4, 5, 6, 7, 8],
           xticks_angle=90,
           yticks=[],
           yticks_angle=90,
           title_size=14,
           title_weight="normal",
           subtitle_size=8,
           subtitle_weight="bold",
           axislabel_size=12,
           axislabel_weight="normal",
           ticklabel_size=8)
```
