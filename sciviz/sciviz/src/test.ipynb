```python
import matplotlib.pyplot as plt
import seaborn as sns
from point import point
from histogram import histogram
from sciviz import sciviz
from theme import theme
```

```python
iris = sns.load_dataset("iris")

# Create the visualization with faceting by species
ax = sciviz(iris, height=6, aspect=1)
ax = histogram(ax=ax, data=iris, x="sepal_length", alpha = 0.8, bins=10, fill=True, binwidth=0.5, palette="deep")
plt.show()
```
