```python
import matplotlib.pyplot as plt
import seaborn as sns
from point import point
from histogram import histogram
from boxplot import boxplot
from bar import bar
from jitter import jitter
from line import line
from sciviz import sciviz
from theme import theme
from violin import violin
from pie import pie
from venn import venn
from euler import euler
from heatmap import heatmap
```

```python
iris = sns.load_dataset("iris")
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')

# Create the visualization with faceting by species
ax = sciviz(height=5, aspect=1)
ax = point(ax=ax, data=iris, x="sepal_length", y="sepal_width", color='species')
plt.show()
```

```python
iris = sns.load_dataset("iris")
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
col_annot_names = 'flowers part'
col_annot = ['sepal_length', 'petal_length', 'petal_width', 'sepal_width']
col_pal = ['red', 'red', 'blue', 'blue']
#species = iris.pop("species")  
ax = sciviz(height=6, aspect=1)
ax = heatmap(ax, iris, cluster_rows=True, show_cbar=True, xticks=True, cbar_ticks=[2, 4, 6], cbar_lim=[0,8], cluster_cols=False, row_annot=['species', 'size'], row_pal=['jama', 'deep'], col_annot=col_annot, col_pal=col_pal, col_annot_names=col_annot_names)
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights['season'] = flights['month'].apply(lambda x: 'winter' if x in ['Jan', 'Feb', 'Mar', 'Oct', 'Nov', 'Dec'] else 'summer')

# Create the visualization with faceting by species
ax = sciviz(height=5, aspect=1)
ax = line(ax=ax, data=flights, x='year', y='passengers', color='season')
user_labels = {'Season': ['winter', 'summer']}
plt.show()
```
