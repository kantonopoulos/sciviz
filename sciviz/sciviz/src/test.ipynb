```python
import matplotlib.pyplot as plt
import seaborn as sns
from point import point
from histogram import histogram
from boxplot import boxplot
from bar import bar
from jitter import jitter
from line import line
from sciviz import sciviz
from theme import theme
from violin import violin
```

```python
flights = sns.load_dataset("flights")
flights['season'] = flights['month'].apply(lambda x: 'winter' if x in ['Jan', 'Feb', 'Mar', 'Oct', 'Nov', 'Dec'] else 'summer')

# Create the visualization with faceting by species
ax = sciviz(flights, height=5, aspect=1)
ax = line(ax=ax, data=flights, x='year', y='passengers', color='season', errorbar=('ci', 95), errorbar_style='bars', markers=True, alpha = 0.8, format_labels=True, show_legend=True)
user_labels = {'Season': ['winter', 'summer']}

'''ax = theme(ax, 
           title="Flights Dataset", 
           subtitle="Passengers over time", 
           theme="ticks", 
           xlab="Time", 
           ylab="P/F",
           title_size=14,
           title_weight="normal",
           subtitle_size=11,
           subtitle_weight="bold",
           axislabel_size=12,
           axislabel_weight="normal",
           ticklabel_size=8,
           legend=True,
           legend_labels=user_labels, 
           legendtitles_size=11, 
           legendtitles_weight="normal", 
           legendlabels_size=10, 
           legendlabels_weight="normal",
           legend_pos='side')'''
plt.show()
```
