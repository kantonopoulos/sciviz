```python
import matplotlib.pyplot as plt
import seaborn as sns
from point import point
from sciviz import sciviz
from legend import customize_legend
```

```python
iris = sns.load_dataset("iris")

# Create the visualization with faceting by species
g = sciviz(iris, height=4, aspect=1)
point(iris, x="sepal_length", y="sepal_width", color="species", style="species", palette="deep", ax=g)
plt.show()
```

```python
iris = sns.load_dataset("iris")

# Create the visualization with faceting by species
ax = sciviz(iris, cols="species", height=4, aspect=1, col_wrap=3)
ax = point(iris, x="sepal_length", y="sepal_width", color="species", palette="deep", ax=ax)
```
